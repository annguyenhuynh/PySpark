```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType
```

```python
# Create dataframe

emp = spark.createDataFrame(emp_data,schema=emp_schema)
emp.show()
emp.printSchema()
```

```python
# Check number of partitions

emp.rdd.getNumPartitions()
```

```python
# Get employee data where salary is greater than 50000
emp_salary = emp.where("salary > 50000.0")
```

```python
emp_salary.rdd.getNumPartitions()
```

```python
emp_salary.write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv("output/emp")
```
