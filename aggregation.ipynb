```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType, DataType
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("test") \
    .getOrCreate()
```

```python
df = spark.read.option("header","true").csv("output/emp")
```

```python
data_sorted = df.orderBy(col("salary").desc())
data_sorted.show()
```

```python
# Count
emp_count = df.groupBy("department_id").agg(count("employee_id").alias("total_dept_count"))
emp_count.show()
```

```python
# Sum
emp_sum= df.groupBy("department_id").agg(sum("salary").alias("total_dept_salary"))
emp_sum.show()
```

```python
# Avg with having clause
emp_avg = df.groupBy("department_id").agg(avg("salary").alias("avg_dept_salary")).where("avg_dept_salary > 50000.0" )
emp_avg.show()
```
