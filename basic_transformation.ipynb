```python
# Correct and clean way to cast data
df = data.select(
	col("employee_id").cast("int").alias("emp_id"),
	col("department_id").cast("int").alias("dpt_id"),
	col("name"),
	col("age").cast("int"),
	col("gender"),
	col("salary").cast("double"),
	col("hired_date").cast("date")
)
```

```python
# Add a tax column to emp_casted_1
df_taxed = df.withColumn("tax",col("salary") * 0.2)
df_taxed.show()
```

```python
# Using lit (literal) to columns containing constant values
df_new_cols = df_taxed.withColumn("ColumnOne",lit(1)).withColumn("ColumnTwo", lit(2))
df_new_cols.show(5)
```

```python
# Rename columns
rename = df_new_cols.withColumnRenamed("ColumnOne", "Col_1").withColumnRenamed("ColumnTwo", "Col_2")
rename.show(5)
```

```python
rename=rename.drop("Col_1", "Col_2")
rename.show()
```

```python
# Add multiple columns
df_final = df_new_cols.drop("ColumnOne","ColumnTwo")
new_cols = {
	"fee": col("salary") * 0.01,
	"col1": lit(1),
	"col2": lit(2)
}

df_final=df_final.withColumns(new_cols)

df_final.show()
```
